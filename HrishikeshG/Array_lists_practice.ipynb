1) Rotate an array by p positions - Counterclockwise

```python
def rotate(arr, pos):
  l = len(arr)
  pos = pos % l
  arr = arr[pos:] + arr[:pos]
  return arr

if __name__ == "__main__":
  arr  = input("enter an array:")
  arr = [int(i) for i in arr.split()]
  pos = int(input("enter the position you want to shift from"))
  final_arr = rotate(arr, pos)
  print(final_arr)
```

3) Given an array arr[] of size n, the task is to rearrange it in alternate positive and negative manner without changing the relative order of positive and negative numbers.

```python
def rearrange(a):
  l = len(a)
  positive = []
  negative = []
  new = []
  for i in a:
    if i > 0 or i == 0:
      positive.append(i)
    else:
      negative.append(i)
  print("positive list is", positive)
  print("negative list is", negative)
  j = 0
  while j < len(positive) and j < len(negative):
    new.append(positive[j])
    new.append(negative[j])
    j+=1

  #to handle extra positive and negative values
  new.extend(positive[j:])
  new.extend(negative[j:])
  for k in range(len(new)):
    a[k] = new[k]

  return a


if __name__ == "__main__":
  arr = [2,-11,-10, 0, 4,-32,-21]
  rearrange(arr)
  print("new rearranged array is ", arr)
```
